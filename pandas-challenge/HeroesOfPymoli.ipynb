## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Creating bins for ages
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999]

# Categorizing players using age range
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Deducing players count by age and inputing into age bins/range and making final count per age
plyr_cnt["Age Ranges"] = pd.cut(plyr_cnt["Age"], age_bins, labels=bin_labels)
age_dem_tot = plyr_cnt["Age Ranges"].value_counts()

# Calculating percentage of players by age ranges
age_dem_perc = age_dem_tot/num_plyrs*100

# Setting up output dataframe
age_dem = pd.DataFrame({"Total Counts": age_dem_tot, 
                       "Percentage of Players": age_dem_perc})

# Formating output dataframe for final output
age_dem["Percentage of Players"] = age_dem["Percentage of Players"].map("{:.2f}%".format)

# Sorting output dataframe 
age_dem = age_dem.sort_index()

# Calling age demographics output dataframe
age_dem
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Computing to sort for most profitable items by purchase count in descending order
purc_cnt = purchase_data.loc[:,["Item ID", "Item Name", "Price"]] 

#Calculating purchase count; grouped by item id and item name indexed to price
pop_purc_cnt = purc_cnt.groupby(["Item ID", "Item Name"]).count()["Price"]

#Calculating purchase count; grouped by item id and item name indexed to price
pop_purc_tot = purc_cnt.groupby(["Item ID", "Item Name"]).sum()["Price"]

# Calculating average item price
pop_purc_price = purc_cnt.groupby(["Item ID", "Item Name"]).mean()["Price"]

# Creating dataframe for output
pop_tot_data = pd.DataFrame({"Purchase Count": pop_purc_cnt, 
                             "Item Price": pop_purc_price,                                                                       
                             "Total Purchase Value": pop_purc_tot})

# Sorting dataframe by purchase count per most popular items in descending order
pop_tot_data = pop_tot_data.sort_values(["Purchase Count"], ascending = False)

# Computing output dataframe per top 5 popular items
pop_tot_data["Item Price"] = pop_tot_data["Item Price"].map("${:,.2f}".format)
pop_tot_data["Total Purchase Value"] = pop_tot_data["Total Purchase Value"].map("${:,.2f}".format)

# Outputing most popular items dataframe
pop_tot_data.head()
```

## Most Profitable Items

```python
# Repeating above table to sort for most profitable items by total purchase value in descending order
purc_cnt = purchase_data.loc[:,["Item ID", "Item Name", "Price"]] 

#Calculating purchase count; grouped by item id and item name indexed to price
pop_purc_cnt = purc_cnt.groupby(["Item ID", "Item Name"]).count()["Price"]

# Calculating total purchase value
pop_purc_tot = purc_cnt.groupby(["Item ID", "Item Name"]).sum()["Price"]

# Calculating average item price
pop_purc_price = purc_cnt.groupby(["Item ID", "Item Name"]).mean()["Price"]

# Creating dataframe for output
pop_tot_data = pd.DataFrame({"Purchase Count": pop_purc_cnt, 
                             "Item Price": pop_purc_price,                                                                         
                             "Total Purchase Value": pop_purc_tot})

# Sorting dataframe by total purchase value per most popular items in descending order
pop_tot_data = pop_tot_data.sort_values(["Total Purchase Value"], ascending = False)

# Creating output dataframe format for presentation
pop_tot_data["Item Price"] = pop_tot_data["Item Price"].map("${:,.2f}".format)
pop_tot_data["Total Purchase Value"] = pop_tot_data["Total Purchase Value"].map("${:,.2f}".format)

# Outputing most profitable dataframe
pop_tot_data.head()
```
